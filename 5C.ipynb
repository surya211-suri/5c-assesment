```python
import pandas as pd
from datetime import datetime
```

```python
df = pd.read_csv("/content/sales_data_1.csv")
```

```python
df
```

```python
# Convert the "Date" column to a datetime object and extract the day of the week
df['Date'] = pd.to_datetime(df['Date'], format='%d/%m/%y')
df['DayOfWeek'] = df['Date'].dt.day_name()
```

```python
df['Profit'] = (df['Selling price'] - df['Buying price']) * df['Quantity sold']
```

```python
# 1. Calculate the overall gross margin for the business
total_revenue = (df['Selling price'] * df['Quantity sold']).sum()
total_cost = (df['Buying price'] * df['Quantity sold']).sum()
gross_margin = (total_revenue - total_cost) / total_revenue

# 2. Find the most profitable vendor
vendor_profit = df.groupby('Vendor')['Profit'].sum()
most_profitable_vendor = vendor_profit.idxmax()

# 3. Find the least profitable customer
customer_profit = df.groupby('Customer')['Profit'].sum()
least_profitable_customer = customer_profit.idxmin()

# 4. Find the most profitable day of the week
day_profit = df.groupby('DayOfWeek')['Profit'].sum()
most_profitable_day = day_profit.idxmax()

# 5. Find the least profitable day of the week
least_profitable_day = day_profit.idxmin()

print(f"1. Overall Gross Margin: {gross_margin * 100:.2f}%")
print(f"2. Most Profitable Vendor: {most_profitable_vendor}")
print(f"3. Least Profitable Customer: {least_profitable_customer}")
print(f"4. Most Profitable Day of the Week: {most_profitable_day}")
print(f"5. Least Profitable Day of the Week: {least_profitable_day}")
```
